```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("raw/NI School List 2022.csv", 
                 names=["id_raw", "name", "address1", "address2", "address3", "city",
                        "county", "postcode", "phone", "status", "type", "open", "blank", "email"])
```

```python
df = df.drop(columns=["address1", "address2", "address3", "phone", "open", "blank", "email"])
```

```python
df.insert(0, "id", df["id_raw"].map(lambda x: x[:3]+x[4:]).astype("int"))
```

```python
df.to_csv("csv/schools.csv", index=False)
```
